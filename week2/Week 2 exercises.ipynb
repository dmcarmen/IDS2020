## Exercise 1 - Titanic

```python
import pandas as pd
import numpy as np
```

```python
#1
train = pd.read_csv('train.csv')

train
```

```python
#2
train.pop("Name")
train.pop("Ticket")
train.pop("Embarked")

train
```

```python
#3
import math
CLetter = []

for item in train["Cabin"]:
    if type(item) == str:
        CLetter.append(item[0])
    else:
        CLetter.append(float('nan'))

train["Cabin"] = CLetter

train
```

```python
#4
train["Cabin"] = train["Cabin"].astype("category")

train
```

```python
#5
train["Age"] = train["Age"].fillna(train["Age"].mean())
train["Cabin"] =train["Cabin"].fillna(train["Cabin"].mode()[0])
train["Fare"] = train["Fare"].fillna(train["Fare"].mean())

train
```

```python
#6
train.to_csv("clean_train.csv")
train.to_json("clean_train.json", orient="records")
```

The average values from this columns are the same for both groups (survivors and non-survivors), for example, the median age was approximately 30, which was the value we filled that column with. Therefore, the median from this column gives almost no information (and the data in general).

Though, if we ignore the missing data, we get more accurate info but less of it, so the results might not be as good. 
On one hand this kind of approach could give you a better understanding of how influent are the columns with missing data. This way, data will not be corrupted by mean or mode. On the other hand, it will give you less information about other columns.

## Exercise 3 - Text data v2.0

```python
#1
import operator

pos = open("pos.txt","r")
neg = open("neg.txt","r")

pos_dic = {}
for line in pos.read().splitlines():
    for word in line.split():
        if word in pos_dic:
            pos_dic.update({word: pos_dic.get(word) + 1})
        else:
            pos_dic.update({word: 1})
        
neg_dic = {}
for line in neg.read().splitlines():
    for word in line.split():
        if word in neg_dic:
            neg_dic.update({word: neg_dic.get(word) + 1})
        else:
            neg_dic.update({word: 1})

pos_list = sorted(pos_dic.items(), key=operator.itemgetter(1), reverse=True)
neg_list = sorted(neg_dic.items(), key=operator.itemgetter(1), reverse=True)

pos.close()
neg.close()

print("Most common words in pos.txt\n",pos_list[0:20])
print()
print("Most common words in neg.txt\n",neg_list[0:20])
```

```python
from sklearn.feature_extraction.text import *
```

```python
#2-3
pos = open("pos.txt","r")
neg = open("neg.txt","r")

pos_vectorizer = TfidfVectorizer()
neg_vectorizer = TfidfVectorizer()

pos_vectors = pos_vectorizer.fit_transform(pos.read().splitlines())
neg_vectors = neg_vectorizer.fit_transform(neg.read().splitlines())

print("TFIDF pos")
pos_weights = np.asarray(pos_vectors.mean(axis=0)).ravel().tolist()
pos_weights_df = pd.DataFrame({'term': pos_vectorizer.get_feature_names(), 'weight': pos_weights})
pos_by_weights = pos_weights_df.sort_values('weight',ascending=False)
pos_words = pos_by_weights['term'].values.tolist()[:20]
print(pos_words)

print("TFIDF neg")
neg_weights = np.asarray(neg_vectors.mean(axis=0)).ravel().tolist()
neg_weights_df = pd.DataFrame({'term': neg_vectorizer.get_feature_names(), 'weight': neg_weights})
neg_by_weights = neg_weights_df.sort_values('weight',ascending=False)
neg_words = neg_by_weights['term'].values.tolist()[:20]
print(neg_words)
```

```python
pos_list_words, dummy = map(list, zip(*pos_list[:20]))
neg_list_words, dummy = map(list, zip(*neg_list[:20]))

print("Union of most commn words and words with the highest TF/IDF score")
union_neg = set().union(pos_list_words,pos_words)
union_pos = set().union(neg_list_words,neg_words)
print("For positive reviews:", len(union_neg))
print("For negative reviews:", len(union_pos))
print("For both reviews:", len(set().union(union_neg,union_pos)))
```

The words are almost the same as the most common words.In that sense, TF/IDF didn't work as expected. Also, we notice that we can not distinguish the reviews by the most common words (as there are a lot of word shared between both groups).
